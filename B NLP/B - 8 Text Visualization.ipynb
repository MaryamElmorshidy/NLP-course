```python
import spacy
from spacy import displacy
```

```python
nlp=spacy.load("en_core_web_lg")
```

```python
doc1 = nlp(u'Apple is going to build a U.K. factory for $6 million.')
displacy.render(doc1,style="dep",jupyter=True,options={"distance":80})
```

```python
doc2 = nlp("""In ancient Rome, some neighbors live in three adjacent houses. In the center is the house of Senex, who lives there with wife Domina, son Hero, and several slaves, including head slave Hysterium and the musical's main character Pseudolus. A slave belonging to Hero, Pseudolus wishes to buy, win, or steal his freedom. One of the neighboring houses is owned by Marcus Lycus, who is a buyer and seller of beautiful women; the other belongs to the ancient Erronius, who is abroad searching for his long-lost children (stolen in infancy by pirates). One day, Senex and Domina go on a trip and leave Pseudolus in charge of Hero. Hero confides in Pseudolus that he is in love with the lovely Philia, one of the courtesans in the House of Lycus (albeit still a virgin).""")
sent_spans=list(doc2.sents)
displacy.render(sent_spans, style="dep",options={'distance': 80})
```

```python
from IPython.core.display import display,HTML
html = displacy.render(doc1, style="dep")
display(HTML(html))
```

```python
from pathlib import Path
sent_ = ["This is an example.", "This is another one."]
for sent in sent_:
    doc=nlp(sent)
    svg=displacy.render(doc, style="dep", jupyter=False)
    file_name="-".join([x.text for x in doc if not x.is_punct])+"svg"
    #print(file_name)
    out_path=Path(file_name)
    #print(out_path)
    out_path.open("w",encoding="utf-8").write(svg)
```
